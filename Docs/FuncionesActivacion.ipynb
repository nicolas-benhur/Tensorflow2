```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = tf.constant(np.arange(-5,5,0.1), dtype = tf.float32)
b = tf.keras.activations.exponential(a)

plt.plot(a,b.numpy())
plt.title('Activations: Exponential')
plt.ylabel('b')
plt.xlabel('a')
plt.show()
```
